### (Bonus) Gunakan metode sederhana seperti (spearman / cramer / pearson correlation / chi square) untuk menentukan signifikan fitur tersebut

```python
#menggunakan spearman method
from scipy.stats import spearmanr

for col in ['Temperature', 'Humidity', 'Wind_Speed', 'Cloud_Cover', 'Pressure']:
    corr, _ = spearmanr(df[col], df['Rain'])
    print(f"Spearman correlation between {col} and Rain: {corr:.2f}")
```

Dari spearman method didapatkan bahwa:<br>
1. Temperature memiliki hubungan yang negatif, artinya semakin tinggi Temperature maka kemungkinan Rain terjadi akan menurun<br>
2. Humidity memiliki hubungan yang positif, artinya semakin tinggi Humidity maka kemungkinan Rain terjadi semakin tinggi<br>
3. Wind Speed tidak memiliki pengaruh terhadap terjadinya Rain<br>
4. Cloud Cover memiliki hubungan yang positif, artinya semakin tinggi Cloud Cover maka kemungkinan Rain terjadi semakin tinggi<br>
5. Pressure hampir tidak memiliki pengaruh terhadap terjadinya Rain<br>

### a) Identifikasi numerical features dalam dataset. Apakah diperlukan feature scaling sebelum proses modeling? Jika iya, gunakan salah satu metode scaling. Jelaskan alasan pemilihan metode tersebut dan aplikasikan scaling pada dataset.

Numerical features dalam dataset adalah Temperature, Humidity, Wind Speed, Cloud Cover, Pressure karena berupa float.<br>

Apakah diperlukan feature scaling sebelum proses modeling?<br> 
=Perlu. Apabila saya akan menggunakan model yang sensitif saya perlu melakukan feature scaling. Namun karena saya masih belum tau model apa yang saya gunakan,saya hanya mengantisipasi dengan melakukan feature scaling.<br>

```python
#Feature scaling dengan standardization atau Z-score normalization
from sklearn.preprocessing import StandardScaler

#karena features Wind_Speed dan Pressure tidak dibutuhkan untuk target prediksi model
df = df.drop(['Wind_Speed', 'Pressure'], axis=1)

features_to_scale = ['Temperature', 'Humidity', 'Cloud_Cover']

scaler = StandardScaler()

#merubah features
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

print(df)
```

Alasan saya memilih untuk menggunakan metode standardization adalah metode ini adalah agar model memperlakukan semua fitur setara dan berada pada skala yang sama sehingga meningkatkan akurasi algoritma model. Selain itu koefisien yang dihasilkan dari metode ini mudah untuk dipahami, koefisien yang lebih besar menunjukkan bahwa features tersebut memiliki pengaruh yang lebih besar ke target.

### b) Identifikasi categorical features dalam dataset. Gunakan metode encoding yang sesuai untuk mengubah data categorical menjadi numerical. Tampilkan hasil encoding.

Categorial features dalam dataset ini adalah Rain

```python
#encoding menggunakan one-hot encoder
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_rain = encoder.fit_transform(df[['Rain']])
encoded_rain = pd.DataFrame(encoded_rain.toarray(), columns=encoder.categories_[0])

#Gabungkan data yang telah di-encode dengan DataFrame asli
df = pd.concat([df, encoded_rain], axis=1)

#Hapus kolom 'Rain' asli
df.drop('Rain', axis=1, inplace=True)

print(df)
```

### c) Bagi dataset menjadi training dan test set dengan proporsi yang sesuai. Jelaskan alasan pemilihan proporsi tersebut.

Saya akan mengubah dataset menjadi training set dan test set dengan proporsi 70:30. Alasan saya memilih proporsi tersebut adalah karena proporsi tersebut menurut saya seimbang, model memiliki 70% data untuk mempelajari pola dan karakteristik, dan 30% data digunakan untuk menguji seberapa baik model tersebut. Selain itu 70:30 adalah proporsi umum yang sering digunakan.

# 3. Modelling

### Gunakan model classification yang Anda ketahui untuk membangun model dari dataset tersebut. Model yang dibangun diharapkan merupakan best model dari model-model lain yang telah Anda pilih. Sertakan alasan mengapa Anda menggunakan model classification tersebut dibandingkan model lainnya.

```python
#membangun model dengan model classification random forest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

#Memisahkan features (X) dan target (y)
X = df[['Temperature', 'Humidity', 'Cloud_Cover']]
y = df['rain']  #Target output

#Membagi dataset menjadi training (70%) dan testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

#Melatih model dengan data training
rf_model.fit(X_train, y_train)

# Melakukan prediksi pada data testing
y_pred = rf_model.predict(X_test)
```

Alasan saya memilih model classification random forest adalah model random forest memiliki akurasi yang tinggi dan lebih stabil terhadap noise. Random forest juga memiliki mekanisme internal untuk mengurangi overfitting. Random forest juga dapat menangani berbagai jenis data baik numerik atau kategorikal. Selain itu random forest juga memberi hasil yang baik dengan parameter default.

# 4. Evaluation

### Tampilkan confusion matrix dari model yang telah Anda bangun. Sertakan juga evaluation metrics, seperti accuracy score, f1-score, precision score, dan recall dari model tersebut. Jelaskan arti dari masing-masing hasil evaluasi yang telah didapatkan. Jelaskan apakah model yang dibangun mengalami overfitting?

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score

#Evaluation Metrics
print("Classification Report:")
print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)

f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)

precision = precision_score(y_test, y_pred)
print("Precision Score:", precision)

recall = recall_score(y_test, y_pred)
print("Recall Score:", recall) 

#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
import seaborn as sns  # Import seaborn for heatmap visualization (if not already imported)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['No Rain', 'Rain'], yticklabels=['No Rain', 'Rain'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

1. Precision: menunjukkan seberapa akurat model mengklasifikasikan rain. Nilai 1.0 menunjukkan bahwa setiap kali model memprediksi rain, prediksi selalu benar.<br>
2. Recall: menunjukkan seberapa banyak dari semua sampel rain yang berhasil diidentifikasi oleh model. Nilai 1.0 berarti model berhasil menemukan semua kasuss hujan yang ada dalam data.<br>
3. F1-score: rata-rata harmonik antara precision dan recall. Nilai 1.0 menunjukkan keseimbangan yang sempurna antara precision dab recall.<br>
4. Support: menunjukkan jumlah sampel untuk rain dan no rain
5. Accuracy score: seberapa banyak model memberikan prediksi yang benar. Nilai 1.0 menunjukkan bahwa semua prediksi model benar.<br>
6. Confusion matrix: memberikan visual mengenai kinerja model, menunjukkan apakah tebakan model benar atau salah. 

Menurut saya, model yang dibangun tidak mengalami overfitting karena dari evaluation metrics semuanya memiliki nilai 1.0, artinya model mengklasifikasikan data uji dengan benar. Selain itu dari confusion matrix, menunjukkan bahwa model tidak membuat kesalahan prediksi sama sekali.
